```python
import pandas as pd

data = '2000-01-01-2019-06-21-Iraq.csv'
```

```python
df = pd.read_csv(data)
df.head()
```

```python
for i in df.admin1.unique():
    print(i)
```

```python
df_erbil_ninewa = df[df.admin1.isin(['Erbil', 'Ninewa'])]
for i in df_erbil_ninewa.admin1.unique():
    print(i)
```

```python
for i in df_erbil_ninewa.event_type.unique():
    print(i)
```

```python
for i in df_erbil_ninewa.sub_event_type.unique():
    print(i)
```

```python
df_events = df_erbil_ninewa.groupby(['event_type','sub_event_type']).size().reset_index().rename(columns={0:'count'})
df_events
```
